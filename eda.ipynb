Introduction:
Unemployment is a situation when a person actively searches for a job and is unable to find work. Unemployment indicates the health of the economy.
The unemployment rate is the most frequent measure of unemployment. The unemployment rate is the number of people unemployed divided by the working population or people working under labor.

Why does unemployment occur? 

1. Large population.
2. Lack of vocational skills or low educational levels of the working population.
3. Legal complexities, Inadequate state support, and low infrastructural, financial, and market linkages to small businesses make such enterprises unviable with cost and compliance overruns.
4. Inadequate growth of infrastructure and low investments in the manufacturing sector, hence restricting the employment potential of the secondary sector.
5. The huge workforce of the country is associated with the informal sector because of a lack of required education or skills, and this data is not captured in employment statistics.
6. The main cause of structural unemployment is the education provided in schools and colleges is not as per the current requirements of the industries.

```python

# List of countries to process
countries = [
    'South Africa',
    'Djibouti',
    'Eswatini',
    'West Bank and Gaza',
    'Botswana',
    'Lesotho',
    'Congo, Rep.',
    'Gabon',
    'Namibia',
    'St. Vincent and the Grenadines'
]

# Dictionary to store the means
means = {}

# Loop through each country and calculate the mean of the last 5 years
for country in countries:
    country_data = df[df['Country Name'] == country]
    last_5_years = country_data[['2017', '2018', '2019', '2020', '2021']]
    mean_last_5_years = last_5_years.mean(axis=1)
    means[country] = mean_last_5_years.values[0]

# Print the results
for country, mean_value in means.items():
    print(f"Mean of the last 5 years' values for {country}: {mean_value:.2f}")
```
